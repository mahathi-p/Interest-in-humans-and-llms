```python
import textstat
import os
import pandas as pd
import matplotlib.pyplot as plt
import regex as re
import numpy as np
import seaborn as sns
import scipy.stats as stats
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
def remove_stu_tea(text):
    text = text.replace('teacher:', '')
    text = text.replace('student:', '')
    return text
```

```python
conv_data = pd.read_excel('data/conversations_for_llama_6656.xlsx')
conv_data['proc_text'] = conv_data['proc_text'].apply(remove_stu_tea).apply(lambda x: x.lower())
conv_data
```

```python
def find_complexities(test_df):

    test_df['flesch_reading_ease'] = test_df['proc_text'].apply(textstat.flesch_reading_ease)
    test_df['flesch_kincaid_grade'] = test_df['proc_text'].apply(textstat.flesch_kincaid_grade)
    test_df['smog_index'] = test_df['proc_text'].apply(textstat.smog_index)
    test_df['coleman_liau_index'] = test_df['proc_text'].apply(textstat.coleman_liau_index)
    test_df['automated_readability_index'] = test_df['proc_text'].apply(textstat.automated_readability_index)
    test_df['dale_chall_readability_score'] = test_df['proc_text'].apply(textstat.dale_chall_readability_score)
    test_df['spache_readability'] = test_df['proc_text'].apply(textstat.spache_readability)
    test_df['syllable_count'] = test_df['proc_text'].apply(textstat.syllable_count)
    test_df['lexicon_count'] = test_df['proc_text'].apply(textstat.lexicon_count)
    test_df['difficult_words'] = test_df['proc_text'].apply(textstat.difficult_words)
    test_df['gunning_fog'] = test_df['proc_text'].apply(textstat.gunning_fog)
    test_df['linsear_write_formula'] = test_df['proc_text'].apply(textstat.linsear_write_formula)
    test_df['mcalpine_eflaw'] = test_df['proc_text'].apply(textstat.mcalpine_eflaw)
    test_df['text_standard'] = test_df['proc_text'].apply(lambda x: textstat.text_standard(x, float_output=True))

    

    return test_df

```

```python
complexity_df = find_complexities(conv_data)
complexity_df
```

```python
complexity_df.to_excel('data/readbility_metrics.xlsx', index=False)
```
